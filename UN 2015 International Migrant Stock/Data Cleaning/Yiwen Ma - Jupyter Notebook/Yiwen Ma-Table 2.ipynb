```python
#INF1340 DATA CELANING
import pandas as pd

undata = pd.ExcelFile("UN_MigrantStockTotal_2015.xlsx")

#there are multiple sheets in one excel, we should read them seperately 
#Every sheets has few rows of data background information,for easy computational purpose,i choose skip those rows 
TP = pd.read_excel(undata, 'Table 2',skiprows=14) #TP=Total population
```

```python
#check basic dataset feasture
TP.head()
```

```python
#check basic dataset feasture
TP.info()
```

```python
#Priciple 1:Column names need to be informative, variable names and not values
#rename unnamed columns to make them as imformative as the rest
TP=TP.rename(columns={"Total population of both sexes at mid-year (thousands)":"b1990","Unnamed: 5":"b1995","Unnamed: 6":"b2000","Unnamed: 7":"b2005","Unnamed: 8":"b2010","Unnamed: 9":"b2015",
                        "Total male population at mid-year (thousands)":"M1990","Unnamed: 11":"M1995","Unnamed: 12":"M2000","Unnamed: 13" : "M2005","Unnamed: 14" : "M2010", "Unnamed: 15" : "M2015",
                        "Total female population at mid-year (thousands)":"F1990","Unnamed: 17" : "F1995","Unnamed: 18":"F2000","Unnamed: 19":"F2005","Unnamed: 20":"F2010","Unnamed: 21":"F2015"})
TP
#Principle 3:Variables need to be on cells, not rows and columns
#after rename, columns have some variables in the title row and these are not informative as well
```

```python
#drop the first non-informative rows
TP1 = TP[1:]
TP1
```

```python
#Principle 2:Each column needs to consist of one and only one variable
#columns contains multiple variables need to be melted and rearranged,create temporary columns
TP1 = TP1.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code"],var_name=["years_renamed"],value_name="Total population at mid-year")
TP1
```

```python
#split temporary column to two different columns with one and only variable
TP1 = TP1.assign(Gender = lambda x : x.years_renamed.str[0], Year = lambda x : x.years_renamed.str[1:])
TP1
```

```python
#drop the temporary column years_renamedï¼Œrename other columns to be more informative
TP2=TP1.drop(['years_renamed'], axis=1,inplace=True)
TP2=TP1.replace(to_replace=["b","F","M"],value=["Both","Male","Female"])
TP2
```

```python
#Pinciple 3:Variables need to be on cells, not rows and columns 
#Variables in this table alreay in cells
```

```python
#Principle 4:each table column needs to have a singular data type 
#check the current data stye 
TP2.info()
#there are 5 columns are object and 2 columns are float, split them into two tables by data type+unique combination
```

```python
#clean decimal places to integer, back to original number presentation in the excel sheet
TP2["Sort\norder"]=TP2["Sort\norder"].astype("int64")
TP2["Country code"]=TP2["Country code"].astype("int64")
TP2.head(10)
```

```python
#double check data type, ready to split into tables by data type following pricinple 4
TP2.info()
```

```python
TP3 = TP2.set_index(["Major area, region, country or area of destination","Notes","Total population at mid-year","Gender","Year"])
TP3
```

```python
#setting up Maintable for table 2 excel sheet 
Maintable = pd.DataFrame.from_records(
    columns=["ID","Major area, region, country or area of destination","Notes","Total population at mid-year","Gender","Year"],
    data =[
        (a + 1,b,c,d,e,f,)
        for (a, (b,c,d,e,f,)) in enumerate(TP3.index.unique())
    ],
)
```

```python
Maintable.head(20)
```

```python
#setting up Appendix table for table 2 excel sheet 
Appendix = TP3[["Sort\norder","Country code"]].copy()
Appendix["ID"] = Maintable.set_index(["Major area, region, country or area of destination","Notes","Total population at mid-year","Gender","Year"])
```

```python
Appendix = Appendix.reset_index(drop=True).set_index("ID")
Appendix.head(100)
```

```python
Maintable.set_index(["ID"])
```

```python
#replace missing data with NA for future computational use 
Maintable.replace(to_replace="..",value=pd.NA,inplace=True)
```

```python
#Noticed that in the original table the population numer are in thousands numbers
#for easy computaitonal purpose, will not times all of the numers by 1000
Maintable["Total population at mid-year"]=Maintable["Total population at mid-year"]*1000
Maintable.head(20)
```
