```python
#INF1340 DATA CELANING
import pandas as pd

undata = pd.ExcelFile("UN_MigrantStockTotal_2015.xlsx")

#there are multiple sheets in one excel, we should read them seperately 
#Every sheets has few rows of data background information,for easy computational purpose,i choose skip those rows 
fmsp = pd.read_excel(undata, 'Table 4',skiprows=14) #Female migrants as a percentage
```

```python
#check basic dataset feasture
fmsp.head()
```

```python
#check basic dataset feasture
fmsp.info()
```

```python
#Principle 1:Column names need to be informative, variable names and not values
#rename unnamed columns to make them as imformative as the rest
fmsp=fmsp.rename(columns={"Female migrants as a percentage of the international migrant stock":"F1990","Unnamed: 6":"F1995","Unnamed: 7":"F2000","Unnamed: 8":"F2005","Unnamed: 9":"F2010","Unnamed: 10":"F2015"})
fmsp
#Principle 3:Variables need to be on cells, not rows and columns
#after rename, columns have some variables in the title row and these are not informative as well
```

```python
#drop the first non-informative rows
fmsp= fmsp[1:]
fmsp
```

```python
#Principle 2:Each column needs to consist of one and only one variable
#columns contains multiple variables need to be melted and rearranged,create temporary columns
fmsp1 = fmsp.melt(id_vars=["Sort\norder","Major area, region, country or area of destination","Notes","Country code","Type of data (a)"],var_name=["years_renamed"],value_name="Female migrants as a percentage of the international migrant stock(%)")
fmsp1
```

```python
#split temporary column to two different columns with one and only variable
fmsp1 = fmsp1.assign(Gender = lambda x : x.years_renamed.str[0], Year = lambda x : x.years_renamed.str[1:])
fmsp1
```

```python
#drop the temporary column years_renamed，rename other columns to be more informative
fmsp2=fmsp1.drop(['years_renamed'], axis=1,inplace=True)
fmsp2=fmsp1.replace(to_replace=["F"],value=["Female"])
fmsp2
```

```python
#Pinciple 3:Variables need to be on cells, not rows and columns 
#Variables in this table alreay in cells
```

```python
#Principle 4:each table column needs to have a singular data type 
#check the current data stye 
fmsp2.info()
#there are 6 columns are object and 2 columns are float,split them into two tables by data type+unique combination
```

```python
#clean decimal places to integer, back to original number presentation in the excel sheet
fmsp2["Sort\norder"]=fmsp2["Sort\norder"].astype("int64")
fmsp2["Country code"]=fmsp2["Country code"].astype("int64")
fmsp2.head(10)
```

```python
#double check data type, ready to split into tables by data type following pricinple 4
fmsp2.info()
```

```python
fmsp3 = fmsp2.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","Female migrants as a percentage of the international migrant stock(%)","Gender","Year"])
fmsp3
```

```python
#setting up Maintable for table 4 excel sheet 
Maintable = pd.DataFrame.from_records(
    columns=["ID","Major area, region, country or area of destination","Notes","Type of data (a)","Female migrants as a percentage of the international migrant stock(%)","Gender","Year"],
    data =[
        (a + 1,b,c,d,e,f,g)
        for (a, (b,c,d,e,f,g)) in enumerate(fmsp3.index.unique())
    ],
)
```

```python
Maintable.head(30)
```

```python
#setting up Appendix table for table 3 excel sheet 
Appendix = fmsp3[["Sort\norder","Country code"]].copy()
Appendix["ID"] = Maintable.set_index(["Major area, region, country or area of destination","Notes","Type of data (a)","Female migrants as a percentage of the international migrant stock(%)","Gender","Year"])
```

```python
Appendix = Appendix.reset_index(drop=True).set_index("ID")
Appendix.head(20)
```
